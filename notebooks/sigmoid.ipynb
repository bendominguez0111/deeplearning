```python
import numpy as np
import sympy as sp
from matplotlib import pyplot as plt
```

```python
z = sp.symbols('z')
sigmoid = 1 / (1 + sp.exp(-z))

display(sigmoid)
```

```python
def sigmoid_func(z):
    return 1 / (1 + np.exp(-z))
```

```python
fig, ax = plt.subplots(figsize=(5, 3))

x = np.linspace(-10, 10, 1000)
sigmoid_func_v = np.vectorize(sigmoid_func)
y = sigmoid_func_v(x)

ax.plot(x, y);
```
